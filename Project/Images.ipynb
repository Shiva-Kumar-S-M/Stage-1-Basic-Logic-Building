```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Change this to your folder where PNG images are stored
png_folder = r"D:\archive\images\images_normalized"
```

```python
# Get all PNG images from the folder
png_files = [os.path.join(png_folder, f) for f in os.listdir(png_folder) if f.lower().endswith(".png")]

print(f"Total PNG images found: {len(png_files)}")
```

```python
# Open the first image
img = Image.open(png_files[0])

# Display the image
plt.imshow(img, cmap="gray")
plt.title("Sample PNG Image")
plt.axis("off")
plt.show()
```

```python
# Pick first 3 images
images_to_show = png_files[:3]

plt.figure(figsize=(12, 5))  # Set figure size

for i, file in enumerate(images_to_show):
    img = Image.open(file)
    
    # 1 row, 3 columns, position = i+1
    plt.subplot(1, 3, i + 1)
    plt.imshow(img, cmap="gray")
    plt.title(f"Image {i+1}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Get all PNG images
png_files = [os.path.join(png_folder, f) for f in os.listdir(png_folder) if f.lower().endswith(".png")]

print(f"Total PNG images found: {len(png_files)}")

# Limit to first 10 images only
images_to_show = png_files[:10]

# Grid layout setup
cols = 5   # 5 images per row
rows = (len(images_to_show) // cols) + 1

plt.figure(figsize=(18, 8))

for i, file in enumerate(images_to_show):
    img = Image.open(file)
    
    # Set grid position
    plt.subplot(rows, cols, i + 1)
    plt.imshow(img, cmap="gray")
    plt.title(f"Image {i+1}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

```python
# Change these paths to your images
img1_path = r"D:\archive\images\images_normalized\21_IM-0729-1001-0001.dcm.png"
img2_path = r"D:\archive\images\images_normalized\28_IM-1231-1001.dcm.png"

img1 = Image.open(img1_path).convert("L")  # Convert to grayscale
img2 = Image.open(img2_path).convert("L")

# ✅ Resize img2 to match img1's size
img2 = img2.resize(img1.size)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img1, cmap="gray")
plt.title("Image 1")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2, cmap="gray")
plt.title("Image 2")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Convert images to NumPy arrays
img1_np = np.array(img1)
img2_np = np.array(img2)

# Calculate absolute difference
diff = cv2.absdiff(img1_np, img2_np)

# Highlight differences using thresholding
_, diff_thresh = cv2.threshold(diff, 30, 255, cv2.THRESH_BINARY)

# Display difference image
plt.figure(figsize=(6, 6))
plt.imshow(diff_thresh, cmap="hot")
plt.title("Highlighted Differences")
plt.axis("off")
plt.show()
```
