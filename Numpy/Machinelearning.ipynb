```python
import numpy as np #The average value
speed=[100,120,130,140,210,130,120,]
x=np.mean(speed)
print(x)
```

```python
#median Method---The Mid point value
import numpy as np
speed=[7,85,84,75,86,94,75,48,85,65,42]
x=np.median(speed)
print(x)   #its the value in the middle after sorted all the values
```

```python
#Mode returns the number appears the most
from scipy import stats
speed=[10,20,30,10,20,10,30,10,20,20,20,10]
x=stats.mode(speed)
print(x)
```

```python
#Standard deviation Method 
'''A low standard deviation means that most of the numbers are close to the mean (average) value.

A high standard deviation means that the values are spread out over a wider range.'''

import numpy as np
age=[84,42,15,21,10,23,62,13]
x=np.std(age)
print(x)
```

```python
#variance method  
'''Find the mean,then find difference from the mean,for each diffrence find the square value,
then average of these squared sequences'''
import numpy as np
speed=[10,20,30,40,50,10,20,30]
x=np.var(speed)
print(x)
```

```python
#Percentile method
'''Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than.'''
import numpy as np
speed=[45,65,41,52,45,65,48,47,85,65]
x=np.percentile(speed,48)
print(x)
```

```python
#Data distribution
import numpy as np
a=np.random.randint(10,50,250)
print(a)
```

```python
#Historgram--to visualize the data set we can draw a histogram with the data are collected
import numpy as np
import matplotlib.pyplot as plt

x=np.random.randint(0,5,250)

plt.hist(x,200)
plt.show()
```

```python
#Big data distrubutions
import numpy as np
import matplotlib.pyplot as plt
x=np.random.uniform(0.0,5.0,10000)
plt.hist(x,100)
plt.show()
```

```python
#Normal data distrubutions
import numpy as np
import matplotlib.pyplot as plt
x=np.random.normal(5.0,1.0,100000)
plt.hist(x,100)
plt.show()
```

```python
#Scatter Plot
import numpy as np
import matplotlib.pyplot as plt

x=np.random.normal(5,10,1000)
y=np.random.normal(10,5,1000)

plt.scatter(x,y)
plt.show()
```
