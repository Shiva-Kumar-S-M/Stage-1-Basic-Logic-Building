```python
from matplotlib.widgets import Slider

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)
l = plt.imshow(images[0], cmap="gray")
ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03])

slider = Slider(ax_slider, 'Slice', 0, len(images)-1, valinit=0, valstep=1)

def update(val):
    l.set_data(images[int(slider.val)])
    fig.canvas.draw_idle()

slider.on_changed(update)
plt.show()
```

```python
import os
import numpy as np
import pydicom
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
# Set your dataset folder
dicom_folder = "/home/student/Documents/Medical/series-00000"   # <-- Change if needed

# Get all DICOM files and sort them
files = sorted([f for f in os.listdir(dicom_folder) if f.endswith(".dcm")])

# Load images into a NumPy array
images = []
for f in files:
    path = os.path.join(dicom_folder, f)
    ds = pydicom.dcmread(path)
    images.append(ds.pixel_array)

# Convert to 3D NumPy array
images = np.array(images)
print(f"Loaded {images.shape[0]} slices, each of size {images.shape[1]}x{images.shape[2]}")
# Choose a slice index to display
slice_index = 10  # Change between 0 and len(images)-1

plt.figure(figsize=(6,6))
plt.imshow(images[slice_index], cmap="gray")
plt.title(f"Slice {slice_index}")
plt.axis("off")
plt.show()

fig, ax = plt.subplots(figsize=(6,6))
plt.subplots_adjust(bottom=0.25)

# Show first slice
img_display = plt.imshow(images[0], cmap="gray")
plt.title("Use slider to navigate slices")
plt.axis("off")

# Create slider axis
ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03])
slider = Slider(ax_slider, 'Slice', 0, len(images)-1, valinit=0, valstep=1)

# Update function
def update(val):
    img_display.set_data(images[int(slider.val)])
    fig.canvas.draw_idle()

slider.on_changed(update)
plt.show()




```

```python
import os
import fnmatch
import cv2
from tabulate import tabulate

def get_video_duration(video_path):
    """Returns video duration in seconds."""
    try:
        cap = cv2.VideoCapture(video_path)
        fps = cap.get(cv2.CAP_PROP_FPS)
        frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
        cap.release()
        if fps > 0:
            return frame_count / fps
        else:
            return 0
    except:
        return 0

def scan_videos(root_folder):
    folder_video_data = {}

    # Walk through all folders
    for root, dirs, files in os.walk(root_folder):
        mp4_files = fnmatch.filter(files, "*.mp4")

        if mp4_files:
            video_info = []
            for video in mp4_files:
                path = os.path.join(root, video)
                duration = get_video_duration(path)
                minutes = int(duration // 60)
                seconds = int(duration % 60)
                video_info.append([video, f"{minutes}m {seconds}s"])

            folder_video_data[root] = video_info

    return folder_video_data

def display_results(video_data):
    total_videos = 0
    for folder, videos in video_data.items():
        print(f"\n📂 Folder: {folder}")
        print(tabulate(videos, headers=["Video Name", "Duration"], tablefmt="fancy_grid"))
        total_videos += len(videos)

    print(f"\n✅ Total Videos Found: {total_videos}")

# Change this path to your folder
root_path = "D:\Coursera - Machine Learning Specialization"
video_data = scan_videos(root_path)
display_results(video_data)
```

```python
import os
from pathlib import Path
from moviepy.editor import VideoFileClip

def get_video_duration(file_path):
    """Return video duration in seconds."""
    try:
        clip = VideoFileClip(str(file_path))
        duration = clip.duration
        clip.close()
        return duration
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return 0

def format_duration(seconds):
    """Convert seconds into H:M:S format."""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    return f"{hours:02}:{minutes:02}:{secs:02}"

def scan_videos(root_folder):
    root_path = Path(root_folder)
    total_videos = 0
    total_duration = 0

    print("\n================= VIDEO REPORT =================\n")

    # Walk through all subfolders
    for folder_path, _, files in os.walk(root_path):
        folder_videos = 0
        folder_duration = 0
        folder_path = Path(folder_path)

        # List of mp4 files in this folder
        mp4_files = [f for f in files if f.lower().endswith(".mp4")]

        if mp4_files:
            print(f"\n📁 Folder: {folder_path}")
            print("-" * 60)

            for file in mp4_files:
                file_path = folder_path / file
                duration = get_video_duration(file_path)
                print(f"🎥 {file}  -->  ⏱ {format_duration(duration)}")
                
                folder_videos += 1
                folder_duration += duration

            # Folder summary
            print(f"\n   ▶ Total videos in this folder: {folder_videos}")
            print(f"   ⏳ Total duration: {format_duration(folder_duration)}")
            print("-" * 60)

            total_videos += folder_videos
            total_duration += folder_duration

    # Final summary
    print("\n================= FINAL REPORT =================")
    print(f"📌 Total videos found: {total_videos}")
    print(f"⏳ Total duration: {format_duration(total_duration)}")
    print("================================================\n")

# Replace with your parent folder path
scan_videos("D:\Coursera - Machine Learning Specialization")
```

```python

import cv2
import os
from datetime import timedelta

def get_video_duration(video_path):
    """Get duration of a video using OpenCV."""
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return 0.0
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    duration = frame_count / fps if fps > 0 else 0
    cap.release()
    return duration

def scan_videos_in_directory(root_folder):
    total_videos = 0
    total_duration = 0.0

    # Traverse through all folders
    for folder, subfolders, files in os.walk(root_folder):
        folder_videos = 0
        folder_duration = 0.0

        for file in files:
            if file.lower().endswith(".mp4"):
                video_path = os.path.join(folder, file)
                duration = get_video_duration(video_path)
                folder_videos += 1
                folder_duration += duration

                # Print each video's info
                print(f"📹 Video: {video_path}")
                print(f"   └ Duration: {timedelta(seconds=int(duration))}")

        # If folder has videos, show summary
        if folder_videos > 0:
            print("\n📂 Folder Summary:", folder)
            print(f"   Videos in this folder: {folder_videos}")
            print(f"   Total duration: {timedelta(seconds=int(folder_duration))}\n")

        # Add to overall totals
        total_videos += folder_videos
        total_duration += folder_duration

    # Final summary
    print("📊 Overall Summary")
    print(f"   Total videos found: {total_videos}")
    print(f"   Total duration: {timedelta(seconds=int(total_duration))}")

# Change this to your target folder
root_directory = "D:\Coursera - Machine Learning Specialization"
scan_videos_in_directory(root_directory)
```
