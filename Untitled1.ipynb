```python
from tensorflow.keras.models import load_model
import numpy as np

def inspect_model(model_path):
    try:
        # Load the model
        model = load_model(model_path)
        print("✓ Model loaded successfully!")
        
        # Get model summary
        print("\n=== MODEL SUMMARY ===")
        model.summary()
        
        # Get input and output shapes
        print("\n=== INPUT/OUTPUT SHAPES ===")
        print(f"Input shape: {model.input_shape}")
        print(f"Output shape: {model.output_shape}")
        print(f"Number of layers: {len(model.layers)}")
        
        # Get detailed layer information
        print("\n=== LAYER DETAILS ===")
        for i, layer in enumerate(model.layers):
            print(f"\nLayer {i}: {layer.name} - {layer.__class__.__name__}")
            
            # Safely get input shape
            try:
                if hasattr(layer, 'input_shape'):
                    print(f"  Input shape: {layer.input_shape}")
                elif hasattr(layer, 'get_input_shape_at'):
                    print(f"  Input shape: {layer.get_input_shape_at(0)}")
                else:
                    print(f"  Input shape: Not directly accessible")
            except Exception as e:
                print(f"  Input shape: Error accessing - {e}")
            
            # Safely get output shape
            try:
                if hasattr(layer, 'output_shape'):
                    print(f"  Output shape: {layer.output_shape}")
                elif hasattr(layer, 'get_output_shape_at'):
                    print(f"  Output shape: {layer.get_output_shape_at(0)}")
                else:
                    print(f"  Output shape: Not directly accessible")
            except Exception as e:
                print(f"  Output shape: Error accessing - {e}")
            
            print(f"  Trainable: {layer.trainable}")
            
            # Show number of parameters for layers with weights
            if layer.weights:
                total_params = sum([np.prod(w.shape) for w in layer.get_weights()])
                print(f"  Parameters: {total_params:,}")
        
        # Check if model is compiled
        print("\n=== MODEL COMPILATION INFO ===")
        try:
            print(f"Optimizer: {model.optimizer}")
            print(f"Loss: {model.loss}")
            print(f"Metrics: {model.metrics}")
        except:
            print("Model compilation details not available")
            
    except Exception as e:
        print(f"Error loading model: {e}")

# Run the inspection
inspect_model("CNN_model.h5")
```

```python
import h5py

def inspect_h5_file(file_path):
    try:
        with h5py.File(file_path, 'r') as f:
            print("✓ H5 file opened successfully!")
            
            # Function to recursively print the structure
            def print_structure(name, obj, indent=0):
                indent_str = "  " * indent
                if isinstance(obj, h5py.Group):
                    print(f"{indent_str}Group: {name}")
                elif isinstance(obj, h5py.Dataset):
                    print(f"{indent_str}Dataset: {name}")
                    print(f"{indent_str}  Shape: {obj.shape}")
                    print(f"{indent_str}  Data type: {obj.dtype}")
            
            print("\n=== FILE STRUCTURE ===")
            f.visititems(print_structure)
            
    except Exception as e:
        print(f"Error opening H5 file: {e}")

# Run the inspection
inspect_h5_file("CNN_model.h5")
```

```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
import os
import matplotlib.pyplot as plt

# Load your model
model = load_model("CNN_model.h5")
IMG_SIZE = 256  # Based on your model's expected input

def preprocess_image(img_path):
    """
    Preprocess image exactly like in your Flask app
    """
    # Read image using OpenCV
    img = cv2.imread(img_path)
    if img is None:
        raise ValueError(f"Could not read image file: {img_path}")
    
    # Convert BGR to RGB (OpenCV uses BGR by default)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Resize image to match model input
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    
    # Normalize pixel values
    img = img / 255.0
    
    # Add batch dimension
    img_array = np.expand_dims(img, axis=0)
    
    return img_array

def predict_image(img_path):
    """
    Make a prediction on a single image
    """
    try:
        # Preprocess the image
        img_array = preprocess_image(img_path)
        
        # Make prediction
        prediction = model.predict(img_array)
        print(f"Raw prediction: {prediction}")
        
        # Extract scores
        pneumonia_score = float(prediction[0][0])
        normal_score = float(prediction[0][1])
        
        # Apply your decision logic (reversed based on your feedback)
        if normal_score > 0.5:
            result = "PNEUMONIA"
            confidence = normal_score * 100
        else:
            result = "NORMAL"
            confidence = (1 - normal_score) * 100
        
        return {
            "result": result,
            "confidence": confidence,
            "pneumonia_score": pneumonia_score,
            "normal_score": normal_score,
            "raw_prediction": prediction
        }
        
    except Exception as e:
        return {"error": str(e)}

def test_model_with_images(image_folder, num_images=5):
    """
    Test the model with multiple images from a folder
    """
    # Get list of image files
    image_files = [f for f in os.listdir(image_folder) 
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    # Limit to num_images
    image_files = image_files[:min(num_images, len(image_files))]
    
    results = []
    for img_file in image_files:
        img_path = os.path.join(image_folder, img_file)
        print(f"\nTesting image: {img_file}")
        
        # Make prediction
        prediction = predict_image(img_path)
        
        if "error" in prediction:
            print(f"Error: {prediction['error']}")
            continue
            
        # Display results
        print(f"Result: {prediction['result']}")
        print(f"Confidence: {prediction['confidence']:.2f}%")
        print(f"Pneumonia score: {prediction['pneumonia_score']:.6f}")
        print(f"Normal score: {prediction['normal_score']:.6f}")
        
        results.append({
            "image": img_file,
            **prediction
        })
    
    return results

def visualize_predictions(results, image_folder):
    """
    Create a visualization of the predictions
    """
    num_images = len(results)
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))
    
    if num_images == 1:
        axes = [axes]
    
    for i, result in enumerate(results):
        img_path = os.path.join(image_folder, result["image"])
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        axes[i].imshow(img)
        axes[i].set_title(f"{result['result']}\n"
                         f"Confidence: {result['confidence']:.1f}%\n"
                         f"Pneumonia: {result['pneumonia_score']:.3f}\n"
                         f"Normal: {result['normal_score']:.3f}")
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

# Main testing code
if __name__ == "__main__":
    # Test with images from different folders
    pneumonia_test_folder = r"D:\Project\Chest X-Ray Images(Pneumonia)\chest_xray\test\PNEUMONIA"
    normal_test_folder = r"D:\Project\Chest X-Ray Images(Pneumonia)\chest_xray\test\NORMAL"
    
    print("Testing with Pneumonia images:")
    pneumonia_results = test_model_with_images(pneumonia_test_folder, 3)
    
    print("\n" + "="*50 + "\n")
    
    print("Testing with Normal images:")
    normal_results = test_model_with_images(normal_test_folder, 3)
    
    # Visualize results if you have matplotlib
    try:
        print("\nVisualizing Pneumonia predictions:")
        visualize_predictions(pneumonia_results, pneumonia_test_folder)
        
        print("\nVisualizing Normal predictions:")
        visualize_predictions(normal_results, normal_test_folder)
    except ImportError:
        print("Matplotlib not available. Skipping visualization.")
```
